```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_measure(graphname, measurename, emb_size, name, loc="lower right", setlims=False, ymin=0, ymax=1):
    data_dir = os.path.expanduser("../Graphs/"+graphname)
    sketches = {"rwalk": "RW", "nodesketch": "NS", "minwise": "L0", "l1": "L1", "l2": "L2"}
    all_sum = [0 for _ in range(5)]
    markers = ["X", "o", "s", "P", "D"]
    for i, sketchname in enumerate(sketches):
        vals = []
        for hop in range(1, 5):
            path = data_dir + "/results/" +sketchname + "_" + str(emb_size) + "_hop_" + str(hop) + "_" + \
                   measurename + ".txt"
            with open(path, 'r') as f:
                    line = f.readlines()[0]
                    line = line.split(" ")
                    vals.append(float(line[0]))
                    all_sum[hop] += float(line[0])
        print(vals)
        plt.plot([1,2,3,4], vals, marker=markers[i], markersize=8, linestyle="--", linewidth=1, label=sketches[sketchname])
        plt.xticks([1,2,3,4], ["1", "2", "3", "4"])
    location = loc#"lower right"
#     if all_sum[1] > all_sum[4]:
#         location = "lower left"
    plt.legend(loc=location)
    plt.xlabel("Neighborhood depth", fontsize=13)
    if graphname == 'HomoSapiens':
        graphname = 'PPI'
    plt.title(graphname + " : " + name, fontsize=15)
    if setlims:
        plt.ylim([ymin, ymax])
    name = name.replace(" ", "")
    plt.savefig("/home/koki2/Dropbox/Research/DAMI/figures/" + graphname + "_" + name + ".png")
    plt.show()
                
```

```python
def get_best_approaches(graphname, competitor, measurename, emb_size):
    data_dir = os.path.expanduser("../Graphs/"+graphname)
    sketches = {"rwalk": "RW", "nodesketch": "NS", "minwise": "L0", "l1": "L1", "l2": "L2"}
    
    print("\\begin{table}[!h]")
    print("\\begin{center}")
    print("\\begin{tabular}{| c || c | c |}")
    print("\hline")
    print("hop & Winner & Baseline \\\\")
    for hop in range(1, 5):
        vals = []
        for i, sketchname in enumerate(sketches):
            path = data_dir + "/results/" +sketchname + "_" + str(emb_size) + "_hop_" + str(hop) + "_" + \
                   measurename + ".txt"
            with open(path, 'r') as f:
                line = f.readlines()[0]
                line = line.split(" ")
                vals.append((float(line[0]), sketches[sketchname]))  
           
        vals = sorted(vals, reverse=True)
        vals_dict = {v[1]:v[0] for v in vals}
        
        best = vals[0]
        if vals[0][1] == competitor:
            best = vals[1]
        
        row = str(hop) + " & " + best[1] + " : " + str(np.round(best[0], 3)) + \
        " & " + competitor + " : " + str(np.round(vals_dict[competitor], 3)) +" \\\\"
        print(row)
        print("\hline")
    print("\end{tabular}")
    print("\caption{Comparison of COLOGNE and " + competitor + " for " + measurename + " on " + graphname + ".}")
    print("\label{tab:" + measurename + "_comp}")
    print("\end{center}")
    print("\end{table}")
    print("\n")
```

```python
for gn in ["Cora", "Citeseer", "Pubmed", "HomoSapiens", "Wikipedia", "BlogCatalog"]:
    get_best_approaches(gn, "RW", "w2v_macro_AUC", 10)
```

```python
for gn in ["Cora", "Citeseer", "Pubmed", "HomoSapiens", "Wikipedia", "BlogCatalog"]:
    get_best_approaches(gn, "NS", "F1", 25)
```

```python
def get_best_approaches2(graphnames, competitor, measurename, emb_size):
    
    sketches = {"nodesketch": "NS", "minwise": "L0", "l1": "L1", "l2": "L2"}
    
    print("\\begin{table*}")
    print("\\begin{center}")
    print("\\begin{tabular}{|c || cc || cc || cc |}")
    print("\multicolumn{1}{c}{} & \multicolumn{2}{c}{\large \\texttt{" + graphnames[0] + \
          "} } &  \multicolumn{2}{c}{\large \\texttt{" + graphnames[1] + \
          "}} &  \multicolumn{2}{c}{\large \\texttt{" + graphnames[2] + "}}  \\\\ ")
    print("\hline")
    print("hop & COLOGNE & NodeSketch & COLOGNE & NodeSketch & COLOGNE & NodeSketch \\\\")
    print("\hline")
    print("\hline")
    for hop in range(1, 5):
        row = str(hop) 
        for gn in graphnames:
            data_dir = os.path.expanduser("../Graphs/"+gn)
            vals = []
            for i, sketchname in enumerate(sketches):
                path = data_dir + "/results/" +sketchname + "_" + str(emb_size) + "_hop_" + str(hop) + "_" + \
                       measurename + ".txt"
                with open(path, 'r') as f:
                    line = f.readlines()[0]
                    line = line.split(" ")
                    vals.append((float(line[0]), sketches[sketchname]))  

            vals = sorted(vals, reverse=True)
            vals_dict = {v[1]:v[0] for v in vals}

            best = vals[0]
            if vals[0][1] == competitor:
                best = vals[1]

            row += " & "  + str(np.round(best[0], 3)) + "($" + best[1] + "$)"\
            " & "  + str(np.round(vals_dict[competitor], 3)) 
        row += "\\\\"
        print(row)
        print("\hline")
    print("\end{tabular}")
    print("\caption{Comparison of COLOGNE and " + competitor + " for " + measurename + ".}")
    print("\label{tab:" + measurename + "_comp}")
    print("\end{center}")
    print("\end{table*}")
    print("\n")
```

```python
get_best_approaches2(["Cora", "Citeseer", "Pubmed"], "NS", "emb_micro_AUC", 50)
```
